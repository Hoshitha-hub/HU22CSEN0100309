```python
import matplotlib.pyplot as plt
import numpy as np

def plot_agile_iterative_process():
  phases = [
      "Requirements Analysis",
      "Dataset Creation",
      "Preprocessing",
      "Model Training",
      "Evaluation",
      "Visualisation",
      "Iteration & Feedback"
  ]
  # Calculate angles for each phase in a circle
  angles = np.linspace(0, 2 * np.pi, len(phases), endpoint=False).tolist()
  angles += angles[:1]  # Close the circle

  # Create polar plot
  fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'polar': True})
  ax.set_theta_offset(np.pi / 2)  # Rotate starting point
  ax.set_theta_direction(-1)  # Clockwise direction

  # Add labels for each phase
  for i, phase in enumerate(phases):
    ax.text(angles[i], 1.2, phase, ha='center', va='center', fontsize=10, wrap=True)
  #Draw arrows to show iteration between phases
  for i in range(len(phases)):
    next_index =(i+1)%len(phases)
    ax.annotate(
        "",xy=(angles[next_index],1),xytext=(angles[i],1),
        arrowprops=dict(arrowstyle="->",color="blue",lw=1.5)
    )
  # Configure the plot outside the loop
  ax.set_ylim(0,1.5)
  ax.axis('off')
  plt.title("Agile Iterative Process for Weather modeling",fontsize=14,y=1.1)
  plt.show() # Show the plot only once

#Call the function to display the plot
plot_agile_iterative_process()
```
